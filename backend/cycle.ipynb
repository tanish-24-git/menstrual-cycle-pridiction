```python
import pandas as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
df=pd.read_csv(r'D:\eda\model\model dataset\period - Copy.csv')
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['Income'])  
```

```python
# Feature selection
features = [
    "number_of_peak", "Age", "Length_of_cycle", "Length_of_Leutal_Phase",
    "Length_of_menses", "Unusual_Bleeding", "Height", "Weight", "BMI",
    "Mean_of_length_of_cycle", "Menses_score"
]
X = df[features]
y_cycle_length = df["Length_of_cycle"]
y_ovulation_day = df["Estimated_day_of_ovulution"]
```

```python
# Split the data
X_train, X_test, y_train_cycle, y_test_cycle = train_test_split(X, y_cycle_length, test_size=0.2, random_state=42)
_, _, y_train_ovulation, y_test_ovulation = train_test_split(X, y_ovulation_day, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the encoder
label_encoder = LabelEncoder()

# Fit and transform the Unusual_Bleeding column
df["Unusual_Bleeding"] = label_encoder.fit_transform(df["Unusual_Bleeding"])
```

```python
# Using Pandas get_dummies
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)

# Ensure both train and test sets have the same columns
missing_cols = set(X_train.columns) - set(X_test.columns)
for col in missing_cols:
    X_test[col] = 0  # Add missing columns with default value 0
X_test = X_test[X_train.columns]  # Reorder columns to match training set
```

```python
print("X_train data types:\n", X_train.dtypes)
print("X_test data types:\n", X_test.dtypes)
```

```python
# Train Random Forest Regressor for cycle length prediction
model_cycle = RandomForestRegressor(random_state=42)
model_cycle.fit(X_train, y_train_cycle)
```

```python
# One-hot encode the entire dataset
X = pd.get_dummies(X, drop_first=True)

# Split the data
X_train, X_test, y_train_cycle, y_test_cycle, y_train_ovulation, y_test_ovulation = train_test_split(
    X, y_cycle_length, y_ovulation_day, test_size=0.2, random_state=42
)
```

```python
# Add missing columns to X_test with default value 0
missing_cols = set(X_train.columns) - set(X_test.columns)
for col in missing_cols:
    X_test[col] = 0  # Add missing columns with default value 0

# Remove extra columns in X_test that are not in X_train
extra_cols = set(X_test.columns) - set(X_train.columns)
X_test = X_test.drop(columns=extra_cols)

# Reorder X_test columns to match X_train
X_test = X_test[X_train.columns]
```

```python
# Ensure both train and test sets have the same columns
missing_cols = set(X_train.columns) - set(X_test.columns)
for col in missing_cols:
	X_test[col] = 0  # Add missing columns with default value 0

# Remove extra columns in X_test that are not in X_train
extra_cols = set(X_test.columns) - set(X_train.columns)
X_test = X_test.drop(columns=extra_cols)

# Reorder X_test columns to match X_train
X_test = X_test[X_train.columns]

# Train Random Forest Regressor for ovulation day prediction
model_ovulation = RandomForestRegressor(random_state=42)
model_ovulation.fit(X_train, y_train_ovulation)

# Make predictions
y_pred_cycle = model_cycle.predict(X_test)
y_pred_ovulation = model_ovulation.predict(X_test)
```

```python
# Evaluate the models
mae_cycle = mean_absolute_error(y_test_cycle, y_pred_cycle)
r2_cycle = r2_score(y_test_cycle, y_pred_cycle)

mae_ovulation = mean_absolute_error(y_test_ovulation, y_pred_ovulation)
r2_ovulation = r2_score(y_test_ovulation, y_pred_ovulation)

print(f"Cycle Length Prediction - MAE: {mae_cycle}, R²: {r2_cycle}")
print(f"Ovulation Day Prediction - MAE: {mae_ovulation}, R²: {r2_ovulation}")
```
